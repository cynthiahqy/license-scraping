````python
# define list of csvs to read from run_01, run_02 into pd.DataFrame

from_01 = [*dc.left_only]
print(from_01)

from_02 = []
from_02 = [*dc.right_only, *dc.diff_files, *dc.common_files]
from_02
````

# reading in csv & combining into single table

````python


table_01 = []

for f in files_01:
    with open(f, newline='') as file:
        kreader = file.readlines()
        table_01.extend(kreader)
        #        kreader = csv.reader(file, delimiter=',')
        # for row in kreader:
        #    print(row)
````

````python
table_01 = pd.DataFrame([])

for f in files_01:
    df = pd.read_csv(f, header=None)
    table_01 = table_01.append(df)
````

# Tidying dataframe for export

`ALL_df`: is ALL search results concatenated into a single table

`K_df`: is all K search results, with `search_keys` replacing `callsign_leaseID` column to account for missing values.

Some rows of search results had no value for 'Call Sign/Lease ID', even though the rows are returned on search of specific K-starting callsign_leaseID

````python
ALL_df = (pd.concat([combined_01, combined_02])
          .reset_index(drop=True)
         )
````

````python
ALL_df.columns = scraped_names
````
