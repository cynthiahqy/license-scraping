### fix encoding

```python
## paths
source_file = 'UL201932915978.txt'
decoded_file = 'decoded-UL201932915978.txt'
```

```python
# read file to $content, and fix encoding errors
with open(source_file, 'rb') as f:
    content_b = f.read()
    f.close()

# replace line endings
WINDOWS_LINE_ENDING = b'\r\n'
UNIX_LINE_ENDING = b'\n'
content_b = content_b.replace(WINDOWS_LINE_ENDING, UNIX_LINE_ENDING)

# replace empty columns with NONE values
content_b = content_b.replace(b'||', b'|NONE|').replace(b'||', b'|NONE|')

# decode from bytes to str
content_str = content_b.decode(errors="replace")
```

```python
# write decoded file
with open(decoded_file, 'w') as f:
    f.write(content_str)
    f.close()

with open(decoded_file, 'r') as f:
    print(f.readline())
    f.close()
```

### convert parsed_file to dataframe

```python
N=10000

with open(decoded_file) as f:
    head = [next(f) for x in range(N)]
    f.close()

N_lines = "".join(head)
print(N_lines)
```

```python
data = StringIO(N_lines)
df = pd.read_csv(data, sep="|", header=None, lineterminator="\n")
df.tail()
```
